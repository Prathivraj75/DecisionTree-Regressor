```python
from sklearn.preprocessing import LabelEncoder#importing label encoder from sklearn

label = LabelEncoder()#object creation

data['Address'] = label.fit_transform(data['Address'])#applying label encoder to ServiceStationName	
data['Suburb'] = label.fit_transform(data['Suburb'])
data['Brand'] = label.fit_transform(data['Brand'])
data['FuelCode'] = label.fit_transform(data['FuelCode'])
data['PriceUpdatedDate'] = label.fit_transform(data['PriceUpdatedDate'])
data['ServiceStationName'] = label.fit_transform(data['ServiceStationName'])

```

```python
data.head(10)
```

# DECISIONTREE ALGORTHM

```python

## Creating independent and dependent variable
x=data.drop("Price",axis=1)#independent variable
y=data.Price #dependent variable
```

```python
from sklearn .model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn. tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
```

```python
# Check the training accuracy.

y_pred = dt.predict(x_test)
```

```python
y_test
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
r2score=r2_score(y_test,y_pred)
```

```python

print("\nTest MAE:", mean_absolute_error(y_test, y_pred))
print("Test MSE:", mean_squared_error(y_test, y_pred))
print("Test R²:", r2_score(y_test, y_pred))
```
